````python
import copernicusmarine
````

````python
%%time
ds = copernicusmarine.open_dataset(dataset_id='cmems_mod_med_phy-tem_anfc_4.2km-3D_PT1H-m')
````

````python
import hvplot.xarray
import panel as pn
pn.extension()
````
